3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
from scipy.stats import bernoulli

def bernoulli_pdf(p, x):
    return bernoulli.pmf(x, p)

# Example: Success probability 0.5
print("PDF Value for x=1, p=0.5:", bernoulli_pdf(0.5, 1))

```

6  Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution

```python
import numpy as np

# Generate discrete uniform data
outcomes = np.arange(1, 7)
cdf = np.cumsum([1/6] * 6)

# Plot CDF
plt.step(outcomes, cdf, where='post', color='purple')
plt.title("CDF of Discrete Uniform Distribution")
plt.xlabel("Outcomes")
plt.ylabel("Cumulative Probability")
plt.show()

```

14 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution

```python
def calculate_z(x, mean, std_dev):
    return (x - mean) / std_dev

z = calculate_z(75, 50, 10)
print("Z-Score:", z)
```

15  Implement hypothesis testing using Z-statistics for a sample dataset

```python
from scipy.stats import norm

# Example values
sample_mean = 55
population_mean = 50
std_error = 10 / np.sqrt(30)

z_stat = (sample_mean - population_mean) / std_error
p_value = 1 - norm.cdf(z_stat)

print("Z-Statistic:", z_stat, "P-Value:", p_value)
```

16 Create a confidence interval for a dataset using Python and interpret the resul

```python
data = np.random.normal(100, 15, 30)
mean = np.mean(data)
std_error = np.std(data, ddof=1) / np.sqrt(len(data))
margin = 1.96 * std_error

print(f"Confidence Interval: ({mean - margin}, {mean + margin})")

```

22  Implement Z-score normalization on a dataset and visualize its transformation

```python
data = np.random.normal(50, 10, 100)
z_normalized = (data - np.mean(data)) / np.std(data)

plt.hist(z_normalized, bins=30, color='cyan')
plt.title("Z-Score Normalization")
plt.show()
```
